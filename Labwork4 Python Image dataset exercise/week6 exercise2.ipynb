```python
import os
files = os.listdir('234')
import matplotlib.pyplot as plt
import numpy as np

files = [os.path.join('234', file_i)
 for file_i in os.listdir('234')
 if '.jpg' in file_i]

imgs = [plt.imread(files[file_i])
        for file_i in range(10)]
data = np.array(imgs) 
mean_img = np.quantile(data,0,axis=0) 

plt.imshow(mean_img.astype(np.uint8))
```

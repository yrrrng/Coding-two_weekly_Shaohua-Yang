```python
import random
import operator
import time

OPERATORS = {
    '+': operator.add,
    '-': operator.sub,
    '*': operator.mul,
    '/': operator.truediv,
    '//': operator.floordiv,
    '**': operator.pow,
}

def generate_problem(difficulty, precision):
    num_1 = random.randint(1, difficulty * 10)
    num_2 = random.randint(1, difficulty * 100)
    operator_symbol = random.choice(list(OPERATORS.keys()))
    operation = OPERATORS[operator_symbol]
    answer = operation(num_1, num_2)
    

    answer = round(answer, precision)
    

    lucky_number = random.randint(1, 10)
    if answer == lucky_number:
        answer *= 2
        print(f"Bonus round! The answer is the lucky number {lucky_number}! Double points for this question.")
    
    return num_1, num_2, operator_symbol, answer

def ask_question_with_timer(timeout, difficulty, precision):
    num_1, num_2, operator_symbol, answer = generate_problem(difficulty, precision)
    start_time = time.time()

    while time.time() - start_time < timeout:
        guess = input(f"What is {num_1} {operator_symbol} {num_2}? ")

        if guess.replace('.', '').isdigit() and round(float(guess), precision) == answer:
            print("Correct!")
            return True

        elif guess.isdigit() and float(guess) != answer:
            clue = "higher" if float(guess) < answer else "lower"
            print(f"Incorrect. Try guessing {clue} next time.")

    print("Time's up!")
    print(f"The correct answer was {answer}.")
    return False

def play_game_with_timer(timeout, num_questions):
    print("Welcome to the math quiz game!\n")
    score = 0
    difficulty = select_difficulty()
    precision = 2 if difficulty <= 2 else 1
    for i in range(num_questions):
        if i % 5 == 0 and i > 0:
            difficulty += 1
            precision = 2 if difficulty <= 2 else 1
            print(f"\nIncreasing difficulty to {difficulty}...")
        
        if ask_question_with_timer(timeout, difficulty, precision):
            score += 1
            
    print(f"\nGame over! Your final score is {score}.")
    
def select_difficulty():
    while True:
        print("Please select a difficulty level:")
        print("1. Easy")
        print("2. Medium")
        print("3. Hard")
        choice = input("Enter your choice (1, 2, or 3): ")
        if choice.isdigit() and int(choice) in [1, 2, 3]:
            if choice == '1':
                return 1
            elif choice == '2':
                return 2
            elif choice == '3':
                return 3
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")
play_game_with_timer(10, 10)  
```
